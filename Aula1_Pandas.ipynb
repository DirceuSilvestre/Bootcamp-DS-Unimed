```python
#importando a biblioteca pandas
import pandas as pd
```

```python
# dataframe df
df = pd.read_csv("/content/sample_data/california_housing_train.csv")
```

```python
df
```

```python
titulos = list(df.columns)
novos = ['Longitute', 'Latitude', "Idade", "Comodos", "Quartos", "Pessoas", "Famílias", "RendaMedia", "ValorMedioDasCasas"]
```

```python
#aprendi esse parametro novo aqui do rename, o inplace serve para modificar o original ao invés de retornar o modificado
for i in range(len(titulos)):
  df.rename(columns={titulos[i]: novos[i]}, inplace=True)
```
